```python
for i in a:
    tree=et.parse(data_dir+ '/' + i)
    root=tree.getroot()
    for x in root.iter('bndbox'):
        root1=et.Element('root')
        root1=x
    for w in root1.iter('xmin'):
        xmin.append(w.text) 
    for x in root1.iter('ymin'):
        ymin.append(x.text)
    for y in root1.iter('xmax'):
        xmax.append(y.text)
    for z in root1.iter('ymax'):
        ymax.append(z.text) 
```

```python
def predict(image):
    image = preprocess_image(image.copy())
    #image, scale = resize_image(image)

    boxes, scores, labels = model.predict_on_batch(
    np.expand_dims(image, axis=0)
  )

    #boxes /= scale

    return boxes, scores, labels
```

```python
THRES_SCORE = 0.5

def draw_detections(image, boxes, scores, labels):
    for box, score, label in zip(boxes[0], scores[0], labels[0]):
        if score < THRES_SCORE:
            break

        color = label_color(label)

        b = box.astype(int)
        draw_box(image, b, color=color)

        caption = "{:.3f}".format(score)
        draw_caption(image, b, caption)
```

```python
def show_detected_objects(image_name, boxes, scores, labels):
    img_path = '/kaggle/input/helmet/keras-yolo2-master/train_image_folder/applauding_001.jpg'
  
    image = read_image_bgr(img_path)
    draw = image.copy()
    draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)

    draw_detections(draw, boxes, scores, labels)
    plt.figure(figsize=(15,10))
    plt.axis('off')
    plt.imshow(draw)
    plt.show()
```
